```python
import numpy as np
import pylab as py
import scipy.special as sp
import matplotlib.pyplot as plt


def JJ(K, n):
    return sp.jv(0.5*(n-1), n*K**2/(4 + 2*K**2)) - sp.jv(0.5*(n+1), n*K**2/(4 + 2*K**2))
def A(K, n):
    return n**2 * K**2 / (1 + 0.5*K**2)**2 * (JJ(K, n))**2
K = np.linspace(0.1, 100, 10000)

plt.figure(figsize=(2*4,2.5*2))
#plt.plot(K, A(K, 1))
for n in [i for i in range(1, 21) if i % 2 != 0]:
    plt.loglog(K, A(K, n),  label=r'$n = {}$'.format(n))

plt.xlabel(r'${K}$', fontsize=18, labelpad = 0.0)
y = plt.ylabel(r'$A_{k}(K)$', fontsize=18, labelpad = 0.0, rotation=90)
y.set_rotation(0)
   
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['top'].set_position(('axes',0))
ax.yaxis.set_ticks_position('left')
ax.spines['top'].set_position(('data',0))
ax.xaxis.set_label_coords(0.999, -0.03)
ax.yaxis.set_label_coords(-0.03, 1.0)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.01))

plt.xticks([0.1, 1.0, 10],
           [r'$0.1$', r'$1$', r'$10$'], fontsize=16)

#leg = plt.legend(fontsize=16)#loc='upper left')
# Get the bounding box of the original legend
#bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)
 
#xOffset = 0.2
#bb.x0 += 0.1
#bb.x1 += 0.01
#bb.y0 += 0.
#leg.set_bbox_to_anchor(bb, transform = ax.transAxes)
    

plt.xlim(0, 100)
plt.ylim(0.001, 0.5)
plt.tight_layout()
plt.grid()

a=plt.text(0.2, 0.11, r'$n = 1$', fontsize=16)
a=plt.text(0.32, 0.015, r'$n = 3$', fontsize=16)
a=plt.text(0.39, 0.0022, r'$n = 5$', fontsize=16)


plt.savefig('/home/andrei/Documents/diploma/TexPresent/pic/A_K.pdf')#, bbox_inches='tight')
plt.savefig('/home/andrei/Documents/diploma/Diploma/pic/A_K.pdf')#, bbox_inches='tight')
plt.show()

```
